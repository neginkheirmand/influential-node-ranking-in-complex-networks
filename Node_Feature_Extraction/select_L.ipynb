# Select a Good L for the network

Since this L is probably gonna have to work for all our datasets, lets find a good one

```python
import os
import networkx as nx
import numpy as np
import pandas as pd

directory = "./../datasets/"
graph_info = []
graph_list = []
graph_avg_degrees = []
graph_median_degrees = []

for dirpath, _, files in os.walk(directory):
    for filename in files:
        try:
            if filename.endswith(".edges"):
                file_path = os.path.join(dirpath, filename) 
                graph_list.append(file_path)
                graph = nx.read_edgelist(file_path, comments="%", nodetype=int)
                # is_directed = isinstance(graph, nx.DiGraph)
                num_nodes = graph.number_of_nodes()
                num_edges = graph.number_of_edges()
                degrees = [deg for node, deg in graph.degree()]
                median_degree = np.median(degrees)
                avg_degree = 2 * num_edges / num_nodes if num_nodes > 0 else 0
                graph_avg_degrees.append(avg_degree)
                graph_median_degrees.append(median_degree)
                max_degree = max(degrees) if degrees else 0
                min_degree = min(degrees) if degrees else 0
                
                graph_info.append({
                    "graph G": os.path.splitext(filename)[0],
                    # "Is Directed": is_directed,
                    "#Nodes n": num_nodes,
                    "#Edges e": num_edges,
                    "Max Degree e_max": max_degree,
                    "Min Degree e_min": min_degree,
                    "Avg Degree <e>": avg_degree,
                    "median Degree <e>": median_degree,
                })
        except Exception as e: 
            print(e, f'{filename}')

df = pd.DataFrame(graph_info)
# print(df)
df
```

```python

# Calculate the average (mean)
average_avg = np.mean(graph_avg_degrees)
average_median = np.mean(graph_median_degrees)

# Calculate the median
median_avg = np.median(graph_avg_degrees)
median_median = np.median(graph_median_degrees)

print(f"Average of avg degree of graphs: {average_avg}")
print(f"Average of median degree of graphs: {average_median}")
print(f"- Median of avg degree of graphs: {median_avg}")
print(f"- Median of median degree of graphs: {median_median}")
```
